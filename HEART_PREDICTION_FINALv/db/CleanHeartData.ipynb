```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Initial Read
heart = "heart.csv"

heart_df = pd.read_csv(heart, encoding="ISO-8859-1")
```

```python
heart_df.head(5)
```

```python
#Chest Pain Stats -- FINISHED
cleaned_cp = heart_df[["cp"]]
cleaned_cp.head(10)
```

```python
#Group By to get Counts -- FINISHED
counts_cp = cleaned_cp.groupby('cp').size().reset_index(name='counts').sort_values(by=['cp'], ascending = True)

counts_cp
```

```python
#Export -- FINISHED
counts_cp.to_csv("chestPain.csv", index=False, header=True)
```

```python
#Gather Flouroscopy Stats -- FINISHED
cleaned_flouro = heart_df[["ca"]]
cleaned_flouro.head(10)
```
